```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
data.columns = ["Years", "Salary"]
```

```python
x = data['Years']
y = data['Salary']
```

```python
plt.scatter(x,y)
```

```python
def loss_function(m,b,x,y):
  total_error = 0;
  for i in range(len(x)):
    total_error += (y[i] - (m*x+b))**2
  total_error = total_error/float(len(x))
  return total_error
```

```python
from re import L
def gradiant(m_now, b_now, x, y,L):
  m_gradient = 0
  b_gradient = 0

  n = len(x)

  for i in range(n):
    x_i = x[i]
    y_i = y[i]
    m_gradient += -(2/n) * x_i * (y_i-(m_now*x_i + b_now))
    b_gradient += -(2/n) * (y_i-(m_now*x_i+b_now))

  m = m_now - m_gradient*L
  b = b_now - b_gradient*L

  return m,b
```

```python
m = 0
b = 0
L = 0.0001
epochs = 30000
```

```python
for i in range(epochs):
  m, b = gradiant(m,b,x,y,L)

print(m,b)
```

```python
plt.scatter(x,y,color='red')
y_pred = [m*xi + b for xi in x]
plt.plot(x, y_pred, color='blue')
```
